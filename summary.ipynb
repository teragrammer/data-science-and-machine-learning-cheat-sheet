````python
# 5. Metrics and Reporting
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
````

`What is a Classification Report?` function in scikit-learn generates a detailed text report showing key classification metrics for each class in a classification model. It summarizes the model's performance by providing precision, recall, F1-score, and support for each class.

Key metrics in `classification_report`:
- Precision: The ratio of correctly predicted positive observations to the total predicted positives. It indicates how many selected items are relevant.

- Recall: The ratio of correctly predicted positive observations to all actual positives. It measures the ability of the model to find all relevant cases.

- F1-score: The harmonic mean of precision and recall, providing a balance between the two.

- Support: The number of actual instances for each class in the dataset.

````python
print(classification_report(y_test, y_label))
````

`What is a Confusion Matrix?`
It is a table used for classification problems to compare predicted and actual class labels.

In a binary classification, the confusion matrix has four key values:

- True Positive (TP): Model correctly predicts the positive class.

- True Negative (TN): Model correctly predicts the negative class.

- False Positive (FP): Model incorrectly predicts positive when the actual is negative.

- False Negative (FN): Model incorrectly predicts negative when the actual is positive.

````python
confusion_matrix(y_test, y_label)
````

`accuracy_score` measures the accuracy of a classification model by calculating the proportion of correctly predicted labels out of the total number of predictions made. It returns a float between 0 and 1, where 1 means perfect accuracy.

````python
accuracy_score(y_test, y_label)
````

````python
# 6. Improve a model
for i in range(10, 100, 10):
    model = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {model.score(x_test, y_test) * 100:.2f}%, n_estimator={i}")
````

### Convert Data to Numerical

````python
car_sales = pd.read_csv("data/car-sales-extended.csv")

x = car_sales.drop("Price", axis=1)
y = car_sales["Price"]
````

`ColumnTransformer` in scikit-learn is a versatile tool that allows applying different data transformation functions to different columns of a dataset simultaneously. This is particularly useful when working with heterogeneous data types, for example, applying scaling to numerical features and one-hot encoding to categorical features within the same dataset.

What ColumnTransformer does
- It accepts a list of transformer tuples, where each tuple defines:

    - A name for the transformation step,

    - The transformer object (e.g., StandardScaler, OneHotEncoder),

    - The columns (by name or index) to which this transformer should be applied.

- It applies each specified transformer independently to its respective columns.

- The results of all transformers are concatenated side-by-side to produce a single transformed output.

Types of Transformers in scikit-learn
1. Scaling Transformers
    - Purpose: Normalize or standardize feature values to a common scale.

    - Examples:
    
        - StandardScaler: Scales features to have zero mean and unit variance.
        
        - MinMaxScaler: Scales features to a given range, typically.
        
        - RobustScaler: Uses medians and quantiles, robust to outliers.

2. Encoding Transformers
    - Purpose: Convert categorical variables into numerical formats.
    
    - Examples:
    
        - OneHotEncoder: Converts categorical features into one-hot encoded binary vectors.
        
        - OrdinalEncoder: Encodes categories as integers.

3. Imputation Transformers
    - Purpose: Fill missing values in the dataset.
    
    - Examples:
    
        - SimpleImputer: Imputes missing values with mean, median, most frequent, or a constant value.
        
        - KNNImputer: Uses nearest neighbors to impute missing values.

4. Dimensionality Reduction Transformers
    - Purpose: Reduce the number of features while preserving important information.
    
    - Examples:
    
        - PCA (Principal Component Analysis): Projects data into principal components.
        
        - TruncatedSVD: Similar to PCA but works with sparse matrices.
        
        - SelectKBest: Selects the top-k features based on a statistical test.

5. Polynomial and Interaction Transformers
    - Purpose: Generate polynomial and interaction features.
    
    - Examples:
    
        - PolynomialFeatures: Generates polynomial and interaction terms.

6. Custom Transformers
    - Users can create their own transformers by extending TransformerMixin and implementing fit and transform methods for specialized tasks.

````python
# Turn the categories into numbers
transformer = ColumnTransformer([("one_hot", OneHotEncoder(), ["Make", "Colour", "Doors"])], 
                                remainder="passthrough")

transformed_x = transformer.fit_transform(x)
````

````python
model.score(x_test, y_test)
````

### Fill Missing Data

````python
car_sales_missing = pd.read_csv("data/car-sales-extended-missing-data.csv")
````

````python
# drop all rows of 'Price' column with missing data
car_sales_missing.dropna(subset=["Price"], inplace=True)
````

````python
# split x and y
x = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]
````

````python
# fill missing values with scikit-learn
from sklearn.impute import SimpleImputer

# fill categorical values with 'missing' & numerical values with mean
imputer = ColumnTransformer([
    ("cat_imputer", SimpleImputer(strategy="constant", fill_value="missing"), ["Make", "Colour"]),
    ("door_imputer", SimpleImputer(strategy="constant", fill_value=4), ["Doors"]),
    ("num_imputer", SimpleImputer(strategy="mean"), ["Odometer (KM)"])
])

# transform the data
filled_x = imputer.fit_transform(x)
````

````python
# convert to dataframe
car_sales_filled = pd.DataFrame(filled_x, columns=["Make", "Colour", "Doors", "Odometer (KM)"])
````

````python
# Turn the categories into numbers
transformer = ColumnTransformer([
    ("one_hot", OneHotEncoder(), ["Make", "Colour", "Doors"])
])

transformed_x = transformer.fit_transform(car_sales_filled)
````

````python
# fit model
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2)

model = RandomForestRegressor(n_estimators=100)
model.fit(x_train, y_train)
model.score(x_test, y_test)
````

### Area under the receiver operating characteristic curve (AUC/ROC)

* Area under curve (AUC)
* ROC curve

ROC curves are a comparison of a model true positive rate (tpr) versus a models false positive rate (fpr)

* True positive = model predicts 1 when truth is 1
* False postive = model predicts 1 when truth is 0
* True negative = model predicts 0 when truth is 0
* False negative = model preicts 0 when truth is 1

Resources
* https://www.youtube.com/watch?v=4jRBRDbJemM
* https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html

### Creating a ROC plot

````python
def plot_roc_curve(fpr, tpr):
    """
    Plots a ROC curve give the false positive rate (fpr)
    and true positive rate (tpr) of a model
    """
    # plot roc curve
    plt.plot(fpr, tpr, color="orange", label="ROC")
    # plot line with no predective power (baseline)
    plt.plot([0, 1], [0, 1], color="darkblue", linestyle="--", label="Guessing")

    # customize the plot
    plt.xlabel("False positive rate (fpr)")
    plt.ylabel("True positive rate (tpr)")
    plt.title("Receiver Operating Characteristic (ROC) Curve")
    plt.legend()
    plt.show()

plot_roc_curve(fpr, tpr)
````

### Tuning hyperparameter by hand

Let's make 3 sets, training, validation and test

````python
model.get_params()
````

We're going to try and adjust:

* `max_depth`
* `max_features`
* `min_samples_leaf`
* `min_samples_split`
* `n_estimators`

````python
def evaluate_preds(y_true, y_preds):
    """
    Performs evaluation comparison on y_true labels vs. y_preds labels
    on a classification.
    """
    accuracy = accuracy_score(y_true, y_preds)
    precision = precision_score(y_true, y_preds)
    recall = recall_score(y_true, y_preds)
    f1 = f1_score(y_true, y_preds)
    metric_dict = {"accuracy": round(accuracy, 2), "precision": round(precision, 2), "recall": round(recall, 2), "f1": round(f1, 2)}
    
    print(f"Acc: {accuracy * 100:.2f}%")
    print(f"Precesion: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 score: {f1:.2f}")

    return metric_dict
````

````python
np.random.seed(42)

# shuffle the data
heart_disease_shuffled = heart_disease.sample(frac=1)

# split into x and y
x = heart_disease_shuffled.drop("target", axis=1)
y = heart_disease_shuffled["target"]

# split the data into train, validation & test
train_split = round(0.7 * len(heart_disease_shuffled)) # 70% of data
valid_split = round(train_split + 0.15 * len(heart_disease_shuffled)) # 15% of data
x_train, y_train = x[:train_split], y[:train_split]
x_valid, y_valid = x[train_split:valid_split], y[train_split:valid_split]
x_test, y_test = x[valid_split:], y[valid_split:]

model_1 = RandomForestClassifier()
model_1.fit(x_train, y_train)

# make a prediction
y_preds_1 = model_1.predict(x_valid)

# evaluate the classifier on validation set
model_1_metrics = evaluate_preds(y_valid, y_preds_1)
````

````python
np.random.seed(42)

# create a second classifier with different hyperparameters
model_2 = RandomForestClassifier(n_estimators=5)
model_2.fit(x_train, y_train)

# make a prediction
y_preds_2 = model_2.predict(x_valid)

# evaluate the 2nd classifier
model_2_metrics = evaluate_preds(y_valid, y_preds_2)
````

### Hyperparameter tuning with RandomizedSearchCV

````python
from sklearn.model_selection import RandomizedSearchCV
np.random.seed(42)

grid_1 = {"n_estimators": [10, 100, 200, 500, 1000, 1200],
       "max_depth": [None, 5, 10, 20, 30],
       "max_features": ["log2", "sqrt"],
       "min_samples_split": [2, 4, 6],
       "min_samples_leaf": [1, 2, 4]}

# split into x and y
x = heart_disease_shuffled.drop("target", axis=1)
y = heart_disease_shuffled["target"]

# split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# instantiate model
model = RandomForestClassifier(n_jobs=1)

# setup RandomizedSearchCV
rs_model = RandomizedSearchCV(estimator=model, param_distributions=grid_1, n_iter=50, cv=5) # verbose=2, to view logs

# fit the RandomizedSearchCV version of model
rs_model.fit(x_train, y_train)

# list the best parameters
rs_model.best_params_
````

````python
# make predictions withh the best hyperparameters
rs_y_preds = rs_model.predict(x_test)

# evaluate the predictions
rs_metrics = evaluate_preds(y_test, rs_y_preds)
````

### Hyperparameter tuning with GridSearchCV

````python
from sklearn.model_selection import GridSearchCV
np.random.seed(42)

grid_2 = {"n_estimators": [100, 200, 500], 
        "max_depth": [None], 
        "max_features": ["log2", "sqrt"], 
        "min_samples_split": [6], 
        "min_samples_leaf": [1, 2]}

# split into x & y
x = heart_disease_shuffled.drop("target", axis=1)
y = heart_disease_shuffled["target"]

# split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# instantiate model
model = RandomForestClassifier(n_jobs=1)

# setup GridSearchCV
gs_model = GridSearchCV(estimator=model,
                       param_grid=grid_2,
                       cv=5) # verbose=2, to view logs

# fit the GridSearchCV version of model
gs_model.fit(x_train, y_train)

# list the best parameters
gs_model.best_params_
````

````python
gs_y_preds = gs_model.predict(x_test)

# evaluate the predictions
gs_metrics = evaluate_preds(y_test, gs_y_preds)
````
