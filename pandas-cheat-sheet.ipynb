### Formatting Data and Manipulation

```python
unformattedData = pd.DataFrame({
    "Name": ["BMW", "Toyota", "Nissan"],
    "Price": ["SG$1,000.00", "SG$2,000.50", "SG$5,000.00"],
    "Doors": [5, 2, 4]
})
```

```python
unformattedData["Clean Price"] = unformattedData['Price'].str.replace(r'[^\d.]', '', regex=True).astype(float)
```

```python
unformattedData
```

```python
unformattedData["Name"] = unformattedData["Name"].str.lower()
```

```python
unformattedData
```

```python
unformattedData.drop("Doors", axis=1) # drop entire column
```

### Handling Missing Data

```python
car_sales_missing_data = pd.read_csv("data/car_sales_missing_data.csv")
```

```python
car_sales_missing_data.head()
```

```python
car_sales_missing_data.isnull() # detect missing values
```

```python
car_sales_missing_data.dropna() # drop rows with any missing values
```

```python
car_sales_missing_data.fillna(0) # replace missing values
```

```python
car_sales_missing_data['Sales in thousands'].fillna(car_sales_missing_data['Sales in thousands'].mean()) # replace missing values with mean value
```
