```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Setup and Basic Plotting

```python
data = [280, 713, 252, 690, 467, 546, 861, 220, 887, 340, 106, 145, 294]
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

### Customized Plot

```python
# setup
fig, ax = plt.subplots(figsize=(5, 5)) # figsize = width and height

# plot your data
ax.plot(x, y)

# customize the labels and title
ax.set(title="Simple Plot", 
       xlabel="x-axis", 
       ylabel="y-axis")

# save as image file
fig.savefig("images/simple-plot.png")
plt.show()
```

### Different Plot and Figure Types

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.plot(x, y) # Line Plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.scatter(x, y) # Scatter Plot
plt.show()
```

```python
car_data = {"Toyota": 100, "Honda": 200, "Nissan": 500}
fig, ax = plt.subplots(figsize=(3, 3))
ax.bar(car_data.keys(), car_data.values()) # Bar Plot, barh for horizontal
ax.set(title="Car Prices", ylabel="Prices ($)")
plt.show()
```

```python
hist_data = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(5, 5))
ax.hist(hist_data) # Scatter Plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.boxplot(data) # Boxplot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
ax.pie(data) # Pie Chart
plt.show()
```

### Multiple Subplots

```python
# Option 1
# for plot multiple subplots

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax1.plot(x, y)
ax2.scatter(x, y)
ax3.bar(car_data.keys(), car_data.values())
ax4.hist(hist_data)
plt.show()
```

```python
# Option 2
# for plot multiple subplots

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax[0, 0].plot(x, y)
ax[0, 1].scatter(x, y)
ax[1, 0].bar(car_data.keys(), car_data.values())
ax[1, 1].hist(hist_data)
plt.show()
```

```python
# Try a different dataset

heart_disease = pd.read_csv("data/heart_disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=25)
```

```python
heart_disease.plot.hist(figsize=(30, 30), subplots=True)
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.plot(kind='scatter', x='age', y='chol', c='target')
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)

ax.set_xlim([45, 85]) # set x-axio limit
```

```python
# OO Method from Scratch
fig, ax = plt.subplots(figsize=(10, 5))

scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# customize the title and labels
ax.set(title='Heart Disease and Cholesterol Levels', xlabel='age', ylabel='cholesterol')

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50['chol'].mean(), linestyle="--")
```

```python
# Subplot of age, cholesterol and thalach
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15, 12))

scatter = ax1.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax1.set(title='Heart Disease and Cholosterol Levels', xlabel='age', ylabel='cholesterol')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(over_50['chol'].mean(), linestyle='--')

scatter = ax2.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax2.set(title='Heart Disease and Cholosterol Levels', xlabel='age', ylabel='thalach')
ax2.legend(*scatter.legend_elements(), title='Target')
ax2.axhline(over_50['thalach'].mean(), linestyle='--')

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
# customizing
plt.style.use('ggplot')

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(15, 12))

scatter = ax1.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')
ax1.set(title='Heart Disease and Cholosterol Levels', xlabel='age', ylabel='cholesterol')
ax1.set_xlim([50, 80])
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(over_50['chol'].mean(), linestyle='--')

scatter = ax2.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter')
ax2.set_xlim([50, 80])
ax2.set_ylim([10, 200])
ax2.set(title='Heart Disease and Cholosterol Levels', xlabel='age', ylabel='thalach')
ax2.legend(*scatter.legend_elements(), title='Target')
ax2.axhline(over_50['thalach'].mean(), linestyle='--')

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
